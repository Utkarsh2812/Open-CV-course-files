```python
import numpy as np
import cv2
import  matplotlib.pyplot as plt
```

```python
img = cv2.imread('Dog.jpg')
copyright = cv2.imread('CV_text.png')
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fix_copyright = cv2.cvtColor(copyright, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
plt.imshow(fix_copyright)
```

```python
# Resize image
img1 = cv2.resize(fix_img, (800,800))
img2 = cv2.resize(fix_copyright, (800,800))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

formula = img1 * alpha + img2 * beta + gamma

```python
img_blended = cv2.addWeighted(src1 = img1, alpha = 0.3, src2 = img2, beta = 0.4, gamma = 40)
plt.imshow(img_blended)
```
