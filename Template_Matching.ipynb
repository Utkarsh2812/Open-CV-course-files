```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
horse = cv2.imread("horses.png")
horse_rgb = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)
plt.imshow(horse_rgb)
```

```python
horse_face = cv2.imread("horse_face.png")
horse_face_rgb = cv2.cvtColor(horse_face,cv2.COLOR_BGR2RGB)
plt.imshow(horse_face_rgb)
```

```python
horse_face_rgb.shape
```

```python
temp_h, temp_w, channels = horse_face_rgb.shape
```

```python
result = cv2.matchTemplate(horse_rgb, horse_face_rgb,  cv2.TM_CCORR_NORMED)
```

```python
plt.imshow(result)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

```python
top_left = max_loc
top_left
```

```python
bottom_right = (top_left[0] + temp_w, top_left[1] + temp_h)
bottom_right
```

```python
cted_temp = cv2.rectangle(horse_rgb, top_left, bottom_right, (255,255,255), 5)
plt.imshow(cted_temp)
```

```python
result2 = cv2.matchTemplate(horse_rgb, horse_face_rgb, cv2.TM_SQDIFF_NORMED)
```

```python
plt.imshow(result2)
```

```python
min_val2, max_val2, min_loc2, max_loc2 = cv2.minMaxLoc(result2)
```

```python
top_left2 = min_loc2
```

```python

top_left2
```
