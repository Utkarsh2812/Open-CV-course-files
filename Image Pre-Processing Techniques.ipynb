```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = 255 * np.zeros((600, 600, 3), dtype=np.uint8)

cv2.putText(image, "ROBOTICS", (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 255, 255), 20, cv2.LINE_AA)

# Display the image using matplotlib
plt.imshow(image)
plt.show()
```

For Erosion 

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(image,kernel,iterations = 1)
plt.imshow(erosion)
```

For Dialation

```python
dilation = cv2.dilate(image,kernel,iterations = 1)
plt.imshow(dilation)
```

For Opening

```python
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

For Closing

```python
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
```

For Gradient

```python
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient)
```

For Tophat

```python
tophat = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel)
plt.imshow(tophat)
```

For Blackhat

```python
blackhat = cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, kernel)
plt.imshow(blackhat)
```

IMAGE CONTOURING

```python
image = cv2.imread('Contour_img.jpg')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
```

```python
# Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Find Canny edges
edged = cv2.Canny(gray, 30, 200)
```

```python
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print("Number of Contours found = " + str(len(contours)))
```

```python
# Create a blank image for drawing contours
contour_image = np.zeros_like(image)
```

```python
# Draw all contours
# -1 signifies drawing all contours
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 3)
```

```python
plt.imshow(edged, cmap='gray')
plt.title('Canny Edges After Contouring')
```

```python
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Contours')
```
